```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data=[[[i+j] for i in range(5)] for j in range(10000)]
target=[(i+5) for i in range(10000)]
#remember to normalize before training

data=np.array(data)/10000
target=np.array(target)/10000
```

```python
print(data.shape)
print(target.shape)
```

```python
train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.1)
```

```python
model=Sequential()
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences=False))
model.compile(loss='mae',optimizer='adam',metrics=['mse','mae'])
```

```python
history=model.fit(train_data,train_target,epochs=1000)
```

```python
from matplotlib import pyplot as plt

plt.plot(history.history['loss'])
plt.xlabel('loss')
plt.ylabel('epochs')
plt.show()
```

```python
results=model.predict(test_data[0:5])
print(test_data[0:5].T*10000)
print(results.T*10000)
```
